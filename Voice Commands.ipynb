```python
import speech_recognition as sr
import subprocess as sp
import os
# import spacy
# nlp = spacy.load('en_core_web_sm')
r=sr.Recognizer()
```

```python
def listen(beginmsg):
    with sr.Microphone() as source:
        r.adjust_for_ambient_noise(source, duration=4)
        print(beginmsg)
        audio = r.listen(source)
        print('end')
    try:
        return r.recognize_google(audio)
    except sr.UnknownValueError:
        return 'Try again'
```

```python
def getDomain(i):
    j=''
    for item in i.lower().split():
        if item == 'dot':
            item='.'
        j=j+item
    return j
```

```python
def command():
    h=listen('begin speaking')
    print(h)
    
    if ('make' in h or 'create' in h) and ('directory' in h or 'folder' in h) :
        i= listen('name the folder(s)')
        print(i.lower())
        i,j= sp.getstatusoutput('mkdir {}'.format(i.lower()))
        if i==0:
            print(j)
        else:
            print('Failed')
    elif ('make' in h or 'create' in h) and ('file' in h or 'document' in h) :
        i= listen('name the file(s)')
        print(i.lower())
        sp.getoutput('touch {}'.format(i.lower()))
    
    elif ('remove' in h or 'delete' in h) and ('directory' in h or 'folder' in h) :
        i= listen('name the folder(s)')
        print(i.lower())
        sp.getoutput('rmdir {}'.format(i.lower()))
    elif ('remove' in h or 'delete' in h) and ('file' in h or 'document' in h) :
        i= listen('name the file(s)')
        print(i.lower())
        sp.getoutput('rm {}'.format(i.lower()))
    elif 'list' in h:
        print(sp.getoutput('ls -la'))
    elif 'date' in h or 'time' in h:
        i,j= sp.getstatusoutput('date')
        if i==0:
            print(j)
        else:
            print('Failed')
    elif 'calendar' in h or 'month' in h:
        print(sp.getoutput('cal'))
    elif 'content' in h or 'print' in h or 'open' in h:
        i= listen('name the file(s)')
        print(i.lower())
        print(sp.getoutput('cat {}'.format(i.lower())))
    elif 'concatenate' in h or 'append' in h or 'add' in h:
        i= listen('name the file(s)')
        print(i.lower())
        j= listen('Enter content')
        print(j.lower())
        sp.getoutput('echo {} > {}'.format(j.lower(), i.lower()))
    elif 'back' in h:
        sp.getoutput('cd ..')
        print(sp.getoutput('pwd'))
    elif ('present' in h or 'current' in h) and 'directory' in h:
        print(sp.getoutput('pwd'))
    elif 'memory' in h or 'RAM' in h:
        print(sp.getoutput('free -h'))
    elif 'ping'in h:
        i=listen('Server IP')
        j=getDomain(i)    
        print(j)
        sp.getoutput('touch ping.txt')
        sp.getoutput('ping -c 8 {} > ping.txt 2>&1'.format(j))
        print(sp.getoutput('cat ping.txt'))
    elif 'who is' in h or 'whois' in h:
        i=listen('Domain Name')
        j=getDomain(i)
        print(j)
        print(sp.getoutput('whois {}'.format(j)))
    elif 'website' in h or 'visit' in h :
        i=listen('Domain Name')
        print(j)
        j=getDomain(i)
        print(sp.getoutput('firefox {}'.format(j)))
    elif 'IP' in h and 'private' in h:
        print(sp.getoutput('ifconfig -a'))
    elif ('DNS' in h and 'lookup' in h) or 'dig' in h:
        i=listen('Domain Name')
        j=getDomain(i)
        print(j)
        print(sp.getoutput('dig {}'.format(j)))
        
    

        
    return
```

```python
command()
```

```python
def extract_proper_nouns(doc):
    pos = [tok.i for tok in doc if tok.pos_ == "PROPN"]
    consecutives = []
    current = []
    for elt in pos:
        if len(current) == 0:
            current.append(elt)
        else:
            if current[-1] == elt - 1:
                current.append(elt)
            else:
                consecutives.append(current)
                current = [elt]
    if len(current) != 0:
        consecutives.append(current)
    return [doc[consecutive[0]:consecutive[-1]+1] for consecutive in consecutives]
```
